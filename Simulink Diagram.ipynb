#### Task 2.

The logic for choosing a driving mode is described by the following pseudocode.

```python
if (speed ≥ 100 km/h) OR 
  (speed ≥ 35 km/h AND
   abs(lateral acceleration) ≥ 3 m/s2)
    use mode 2
else
    use mode 1
end
```

You will model the boolean statements first, then add the logic.

Add three Compare to Constant blocks to the model. Set their conditions to ≥ 100, ≥ 35 and ≥ 3. 

Connect the blocks to their appropriate signals (e.g. ≥ 3 is related to acceleration). Remember, right-click + drag branches a signal.
